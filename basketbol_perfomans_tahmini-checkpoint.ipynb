```python
y = summary["Target"]

x = summary.drop(["Name", "Target"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

Eğitim ve test kümelerini ayıralım.

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.fit_transform(X_test)
```

# 1. Lojistik Regresyon Modeli

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

10 Fold Cross Validation ile modeli iyileştirebiliriz.

```python
cv = KFold(n_splits=10, shuffle=True, random_state=1) 
scoreCV = cross_val_score(lr, x_train, y_train, scoring='accuracy', cv = cv)
```

```python
scoreCV
```

```python
np.mean(scoreCV)
```

```python
y_pred = lr.predict(x_test)
metrics.confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

Bütün değişkenleri kullanarak lojistik regresyon modellemesi yaptığımızda yukarıdaki sonuçları elde ediyoruz.

#### Ridge Regression

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
C_param_range = [0.001,0.01,0.1,1,10,100,1000]
scoresCV = []
selection = SelectKBest(f_classif, k = 15).fit(x_train, y_train)
features = list(x.columns[selection.get_support()])    
x_select = scaler.fit_transform(X_train[features])
x_select_test =scaler.fit_transform(X_test[features])
for c in C_param_range:
    lr_ridge = LogisticRegression(C  = c, penalty = 'l2', max_iter = 1000)
    lr_ridge.fit(x_select, y_train)    
    scoreCV = cross_val_score(lr_ridge, x_select, y_train, scoring='accuracy', cv = cv)
    y_pred = lr_ridge.predict(x_select_test)
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)
    recall = metrics.recall_score(y_test, y_pred)
    scoresCV.append([c,np.mean(scoreCV), accuracy, precision, recall])
```

```python
df_ridge = pd.DataFrame(scoresCV,columns=['C (1/lambda)','Validation Accuracy',"Accuracy","Precision","Recall"])
df_ridge
```

#### Lasso Regression

```python
scoresCV = []
for c in C_param_range:
    lr_lasso = LogisticRegression(C=c,penalty='l1',max_iter=1000, solver="liblinear")
    lr_lasso.fit(x_select, y_train) 
    scoreCV = cross_val_score(lr_lasso, x_select, y_train, scoring='accuracy', cv = cv)
    y_pred = lr_ridge.predict(x_select_test)
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)
    recall = metrics.recall_score(y_test, y_pred)
    scoresCV.append([c,np.mean(scoreCV), accuracy, precision, recall])
```

```python
df_lasso = pd.DataFrame(scoresCV,columns=['C (1/lambda)','Validation Accuracy',"Accuracy","Precision","Recall"])
df_lasso
```

# 2. K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

k=3 olarak belirlediğimizde model aşağıdaki tahmin skorunu üretiyor.

```python
knn = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(knn, x_train, y_train, cv = cv, scoring='accuracy')
print(scores.mean())
```

k'nın optimal değerini belirleyebilmek için bir for döngüsü yazabiliriz.

```python
knnscores = {}
for i in range (3, 250):
    knn = KNeighborsClassifier(n_neighbors = i)
    scores = cross_val_score(knn, x_train, y_train, cv = cv, scoring='accuracy')
    knnscores[scores.mean()] = i

plt.plot(knnscores.keys())
plt.xlabel("Neighbour Size")
plt.ylabel("Validation Accuracy")
plt.show()
m = max(list(knnscores.keys()))
print(m, knnscores[m] )
```

Yukarıdaki grafik incelendiğinde k değişkenini yaklaşık olarak 20 ve daha büyük alırsak ortalamada score %70'lere ulaşıyor.  

```python
knn = KNeighborsClassifier(n_neighbors = knnscores[m])
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
metrics.confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

# 3. Random Forest

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators=100, random_state=0, oob_score=True)
bag.fit(x_train, y_train)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```

```python
Oob_Accuracy=[]
for i in np.linspace(start = 50, stop = 500, num = 10):
    clf=RandomForestClassifier(random_state=0,n_estimators=int(i),oob_score=True)
    clf.fit(x_train,y_train)
    Oob_Accuracy.append([i,np.array(clf.oob_score_)])
df = pd.DataFrame(Oob_Accuracy,columns=['Number_of_Trees','Oob Accuracy'])
df
```

```python
Oob_Accuracy=[]
for i in range(1,20):
    clf=RandomForestClassifier(random_state=0,n_estimators=350,max_features=i,oob_score=True)
    clf.fit(x_train,y_train)
    Oob_Accuracy.append([i,np.array(clf.oob_score_)])
df = pd.DataFrame(Oob_Accuracy,columns=['Number_of_Features','Oob Accuracy'])
df
```
